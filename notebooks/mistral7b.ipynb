```python
import json, torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import PeftModel
from src.format_noc import build_prompt

base_id = "mistralai/Mistral-7B-Instruct-v0.2"
adapter_dir = "outputs/mistral7b-noc-switch-qlora/checkpoint-200"

qcfg = BitsAndBytesConfig(load_in_4bit=True, bnb_4bit_quant_type="nf4",
                          bnb_4bit_compute_dtype=torch.float16, bnb_4bit_use_double_quant=True)

tok = AutoTokenizer.from_pretrained(base_id, use_fast=True)
tok.pad_token = tok.eos_token
tok.padding_side = "right"

base = AutoModelForCausalLM.from_pretrained(base_id, quantization_config=qcfg, device_map="auto", torch_dtype=torch.float16)
ft = PeftModel.from_pretrained(base, adapter_dir)
ft.eval()

with open("data/processed_str/valid.jsonl","r",encoding="utf-8") as f:
    ex = json.loads(next(f))
spec = json.loads(ex["spec"])

prompt = build_prompt(spec)
inp = tok(prompt, return_tensors="pt").to(ft.device)

@torch.no_grad()
def gen(m):
    out = m.generate(**inp, max_new_tokens=256, do_sample=False, temperature=0.0,
                     eos_token_id=tok.eos_token_id, pad_token_id=tok.pad_token_id)
    txt = tok.decode(out[0], skip_special_tokens=True)
    return txt[len(prompt):].strip()

print("=== fine-tuned ===")
print(gen(ft))
```

## Sample Evaluation

```python
import json, math

def parse_json_from_text(txt: str):
    txt = txt.strip()
    l = txt.find("{")
    r = txt.rfind("}")
    if l == -1 or r == -1 or r <= l:
        raise ValueError("No JSON object found in text")
    return json.loads(txt[l:r+1])

def ensure_dict(x):
    if isinstance(x, dict):
        return x
    if isinstance(x, str):
        return json.loads(x)
    raise TypeError(f"Expected dict or str, got {type(x)}")

def ensure_switches_dict(sw):
    if isinstance(sw, dict):
        return sw

    if isinstance(sw, str):
        s = sw.strip()

        if s.startswith("{"):
            obj = json.loads(s)
            if "switches" in obj and isinstance(obj["switches"], dict):
                return obj["switches"]
            if all(isinstance(v, dict) and "x" in v and "y" in v for v in obj.values()):
                return obj
                
        if '"switches"' in s and not s.startswith("{"):
            obj = parse_json_from_text("{" + s + "}")
            return obj["switches"]

        obj = parse_json_from_text(s)
        if isinstance(obj, dict):
            if "switches" in obj and isinstance(obj["switches"], dict):
                return obj["switches"]
            return obj

    raise TypeError(f"Unsupported switches type: {type(sw)}")

def extract_pts_from_switches(sw_dict):
    pts = []
    for name, v in sw_dict.items():
        pts.append((name, float(v["x"]), float(v["y"])))
    return pts

def in_blockage(x, y, blockages):
    for bid, b in blockages.items():
        bx, by = float(b["x"]), float(b["y"])
        bw, bh = float(b["width"]), float(b["height"])
        if bx <= x <= bx + bw and by <= y <= by + bh:
            return bid
    return None

def greedy_match_mean_l2(pred_pts, gt_pts):
    if not pred_pts or not gt_pts:
        return None

    used = set()
    dists = []
    for px, py in pred_pts:
        best_j, best_d = None, 1e18
        for j, (gx, gy) in enumerate(gt_pts):
            if j in used:
                continue
            d = (px - gx) ** 2 + (py - gy) ** 2
            if d < best_d:
                best_d, best_j = d, j
        if best_j is not None:
            used.add(best_j)
            dists.append(math.sqrt(best_d))

    return sum(dists) / len(dists) if dists else None


def eval_one_switch_prediction(spec, pred_text, gt_switches):

    spec = ensure_dict(spec)
    W, H = spec["floorplan_dim"]
    blockages = spec.get("blockages", {})

    # pred switches
    pred_obj = parse_json_from_text(pred_text)
    pred_sw_dict = pred_obj.get("switches", {})
    if not isinstance(pred_sw_dict, dict):
        raise ValueError("Pred JSON has no valid 'switches' dict")

    # gt switches
    gt_sw_dict = ensure_switches_dict(gt_switches)

    pred_sw = extract_pts_from_switches(pred_sw_dict)
    gt_sw   = extract_pts_from_switches(gt_sw_dict)

    # validity checks
    oob, blk = [], []
    for name, x, y in pred_sw:
        if not (0 <= x < W and 0 <= y < H):
            oob.append((name, x, y))
        bid = in_blockage(x, y, blockages)
        if bid is not None:
            blk.append((name, x, y, bid))

    pred_pts = [(x, y) for _, x, y in pred_sw]
    gt_pts   = [(x, y) for _, x, y in gt_sw]

    return {
        "n_pred_switches": len(pred_sw),
        "n_gt_switches": len(gt_sw),
        "count_error": abs(len(pred_sw) - len(gt_sw)),
        "all_in_bounds": len(oob) == 0,
        "all_in_free_space": len(blk) == 0,
        "oob_list": oob,
        "in_blockage_list": blk,
        "mean_matched_l2": greedy_match_mean_l2(pred_pts, gt_pts),
    }
```

```python
pred_text = gen(ft)           

report = eval_one_switch_prediction(
    spec=ex["spec"],
    pred_text=pred_text,
    gt_switches=ex["switches"]  
)

print(json.dumps(report, ensure_ascii=False, indent=2))
```
